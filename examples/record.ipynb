```python
from simworld.utils.video_recorder import VideoRecorder
from simworld.utils.vector import Vector
from simworld.communicator.communicator import Communicator
from simworld.communicator.unrealcv import UnrealCV
from simworld.agent.humanoid import Humanoid

import random
```

```python
humanoids = [Humanoid(Vector(random.randint(-600, 600), random.randint(7000, 9000)), Vector(1, 0)) for i in range(3)]
communicator = Communicator(UnrealCV())
```

```python
for humanoid in humanoids:
    communicator.spawn_agent(humanoid, '')
```

```python
camera_ids = [humanoid.camera_id for humanoid in humanoids]
```

```python
for cam_id in camera_ids:
    communicator.unrealcv.set_camera_resolution(cam_id, (1440, 720))
```

```python
images = {humanoid.camera_id: [] for humanoid in humanoids}
```

```python
req_images = communicator.get_camera_observation_multicam(camera_ids, 'lit', 'direct')
```

```python
print(len(req_images))

print(req_images[0].shape)
```

```python
import cv2

# Export req_images to file
for cam_id, img in enumerate(req_images):
    filename = f'camera_{cam_id}_image.png'
    cv2.imwrite(filename, img)
    print(f'Saved image for camera {cam_id} to {filename}')
```
